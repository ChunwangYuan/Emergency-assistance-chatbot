```python
!pip install tflearn
import numpy
import tflearn
import tensorflow
import random
import json
import pickle
```

```python
w1=[]  #words=w1
lb=[] #labels=lb
d_x=[]
d_y=[]

for intent in data["intents"]:
	#print(intent)
	for pattern in intent["patterns"]:
		#print(pattern)
		t=nltk.word_tokenize(pattern) #wrds=t
		#print(wrds)
		w1.extend(t)
		d_x.append(t)
		d_y.append(intent["tag"])

	if intent["tag"] not in lb:
		lb.append(intent["tag"])

#print(w1)
w1=[stemmer.stem(w.lower()) for w in w1 if w != "?"]
#print(w1)
w1=sorted(list(set(w1)))
lb=sorted(lb)
#print(labels)

training=[]
output=[]

out_empty=[0 for i in range(len(lb))]
print(out_empty)
for x, doc in enumerate(d_x):
	b=[]  #bag=b
	#print(doc)
	t=[stemmer.stem(w) for w in doc]
	#print(wrds)
	for w in w1:
		#print(w)
		if w in t:
			b.append(1)
		else:
			b.append(0)
	#print(b)
	output_row = out_empty[:]
	#print(output_row)
	output_row[lb.index(d_y[x])] = 1
	#print(output_row)

	training.append(b)
	output.append(output_row)
training = numpy.array(training)
print(training.shape)
output = numpy.array(output)
print(output.shape)
```

```python
print(w1)
print(lb)
print(d_x)
print(d_y)
```

```python
!pip install tflearn
tensorflow.compat.v1.reset_default_graph()

net = tflearn.input_data(shape=[None, len(training[0])])
net = tflearn.fully_connected(net, 8)
net = tflearn.fully_connected(net, 8)
net = tflearn.fully_connected(net, len(output[0]), activation="softmax")
net = tflearn.regression(net)

model = tflearn.DNN(net)
```

```python
try:
	model.load("model.tflearn")
except:
	model = tflearn.DNN(net)
	model.fit(training, output, n_epoch=1000, batch_size=8, show_metric=True)
	model.save("model.tflearn")
```

```python
def bag_of_words(s,words):
	b=[0 for i in range(len(words))]


	sw=nltk.word_tokenize(s)
	sw=[stemmer.stem(i.lower()) for i in sw]

	for j in sw:
		for i,w in enumerate(words):
			if w==j:
				b[i]=1

	return numpy.array(b)
```

```python
%%time
def chat():
	print("Start Talking with the bot(type quit to stop!")
	while True:
		ipt=input("You: ")
		if ipt.lower()=="quit":
			break
		
		results = model.predict([bag_of_words(ipt,w1)])[0]
		results_index = numpy.argmax(results)
		tag = lb[results_index]
		
		import time
		start = time.process_time()
		
		if results[results_index] > 0.5:
			for tg in data["intents"]:
				if tg['tag'] == tag:
					responses = tg['responses']

			print(random.choice(responses))
			print("\n")

		else:
			print("I didnt get that, try again")
    	
		print(time.process_time()-start)
```

```python
%%time
chat() 
```
